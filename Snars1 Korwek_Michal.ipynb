```python
#task2

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

n = 10
closed = 0

plt.figure(figsize=(8,6))


X=[]
Y=[]

for i in range(n):
    X.append(np.cos(2*np.pi*i/n))
    Y.append(np.sin(2*np.pi*i/n))

plt.scatter(X,Y, c ='red', marker='*', s = 600)
for i in range(n-1):
    plt.plot([X[i],X[i+1]],[Y[i],Y[i+1]], 'black', linewidth =1)
if(closed == 1):
    plt.plot([X[n-1],X[0]],[Y[n-1],Y[0]], 'black', linewidth =1)
plt.show()
```

```python
import pandas as pd
import numpy as np
import turtle
from PIL import Image
#task 5 and 6
n = 6

plt.figure(figsize=(10,8))


X=[]
Y=[]

for i in range(n):
    X.append(np.cos(2*np.pi*i/n))
    Y.append(np.sin(2*np.pi*i/n))

plt.scatter(X,Y, c ='red', marker='*', s =600)

for i in range(n):
    plt.text(X[i]*1.085, Y[i]*1.085, str(i), fontsize=14, fontweight='bold', ha='center', va='center')

deg_out = {i: 0 for i in range(n)}
deg_in = {i: 0 for i in range(n)}
for i in range(n):
    for k in range (n-i-1):
        
        if(np.random.random()>0.5):
            print(str(i)+" "+str(n-k-1))
            
            plt.arrow(X[i],Y[i],X[n-k-1] - X[i],Y[n-k-1] - Y[i], head_width=0.1, head_length=0.12,
                    fc='black', ec='black', alpha=0.7, length_includes_head=True, zorder=2)
            deg_out[i]+=1
            deg_in[n-k-1]+=1
print('deg_out: ')
print(deg_out)
print('deg_in: ')
print(deg_in)


```
