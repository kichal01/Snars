```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import random
```

```python
G_er = nx.erdos_renyi_graph(100, 0.05, seed=42)
avg_degree = []
avg_neigb_degree = []
for node in G_er.nodes():
    knn = list(G_er.neighbors(node))

    ki = len(knn)
    avg_degree.append(ki)
    if ki > 0:
        s = 0
        for neighbor in knn:
            s += len(list(G_er.neighbors(neighbor)))
        result = s/ki
    else: result = 0
    avg_neigb_degree.append(result)
            

fig = plt.figure(figsize=(16,8))
ax = fig.subplots(1, 2)

ax[0].scatter(avg_degree, avg_neigb_degree)

z = np.polyfit(avg_degree, avg_neigb_degree, 1)
p = np.poly1d(z)
ax[0].plot(avg_degree, p(avg_degree), "r--")
ax[0].set_title("ER graph")
ax[0].set_xlabel("degree of node")
ax[0].set_ylabel("average degree of neighbors")

pos = nx.spring_layout(G_er, seed=10396953)
nx.draw_networkx_nodes(G_er,  ax=ax[1], pos = pos, node_size=25)
nx.draw_networkx_edges(G_er,  ax=ax[1], pos = pos, alpha=0.6)
ax[1].set_title("ER graph")

r = nx.degree_assortativity_coefficient(G_er)
print(f"Degree correlation coefficient: {r:.3f}")
```

```python
def switch(G_er):
    u, v = random.sample(list(G_er.nodes()), 2)

    if G_er.has_edge(u, v):
        G_er.remove_edge(u, v)
        print(f"Usunito krawd ({u}, {v})")
    else:
        G_er.add_edge(u, v)
        print(f"Dodano krawd ({u}, {v})")
def edge_switch(G_er):
    u, v = random.sample(list(G_er.edges()), 2) # remove
    # add another

    if G_er.has_edge(u, v):
        G_er.remove_edge(u, v)
        print(f"Usunito krawd ({u}, {v})")
    else:
        G_er.add_edge(u, v)
        print(f"Dodano krawd ({u}, {v})")

def stats(G_er):
    avg_degree = []
    avg_neigb_degree = []
    for node in G_er.nodes():
        knn = list(G_er.neighbors(node))

        ki = len(knn)
        avg_degree.append(ki)
        if ki > 0:
            s = 0
            for neighbor in knn:
                s += len(list(G_er.neighbors(neighbor)))
            result = s/ki
        else: result = 0
        avg_neigb_degree.append(result)
        
    return avg_degree, avg_neigb_degree
```

```python
G_er = nx.erdos_renyi_graph(100, 0.05, seed=42)


fig, axes = plt.subplots(3, 3, figsize=(20, 12))

for i in range(9):  
    avg_degree, avg_neigb_degree = stats(G_er)


    z = np.polyfit(avg_degree, avg_neigb_degree, 1)
    p = np.poly1d(z)
    ax = axes.flat[i]

    x = np.linspace(0,20,40)
    ax.plot(x, p(x), "r--")
    ax.set_title(f"ER graph after {(i*15)} changes")
    ax.set_xlabel("degree of node")
    ax.set_ylabel("average degree of neighbors")
    ax.set_xlim(0,20)
    ax.set_ylim(0,15)

    
    ax.scatter(avg_degree, avg_neigb_degree)

    for j in range(15):
        switch(G_er)

r = nx.degree_assortativity_coefficient(G_er)
print(f"Degree correlation coefficient: {r:.3f}")
```
