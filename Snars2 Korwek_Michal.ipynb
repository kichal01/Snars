```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
G = nx.read_edgelist("data/aves-weaver-social-00.edges")

print(f"Liczba węzłów: {G.number_of_nodes()}")
print(f"Liczba krawędzi: {G.number_of_edges()}")
print(f"Skierowany: {G.is_directed()}")



nx.draw(G, node_size=300, node_color="green", edge_color="gray", with_labels=True)
plt.title("Graph from aves-weaver-social-00.edges file ")
plt.show()
```

```python

G = nx.read_edgelist("data/aves-sparrow-social-2009.edges",data=(("weight", float),))

print(f"Liczba węzłów: {G.number_of_nodes()}")
print(f"Liczba krawędzi: {G.number_of_edges()}")
print(f"Skierowany: {G.is_directed()}")

weights = np.array([d['weight'] for (_, _, d) in G.edges(data=True)])

weights_norm = 5*(weights - weights.min()) / (weights.max() - weights.min())-1

nx.draw(G,width = weights_norm, node_size=140, node_color="skyblue", edge_color="gray", with_labels=True)
plt.title("Graph from aves-sparrow-social-2009.edges file ")
plt.show()
```

```python

G = nx.read_edgelist("data/congress_network/congress.edgelist",data=(("weight", float),))

print(f"Liczba węzłów: {G.number_of_nodes()}")
print(f"Liczba krawędzi: {G.number_of_edges()}")
print(f"Skierowany: {G.is_directed()}")

weights = np.array([d['weight'] for (_, _, d) in G.edges(data=True)])

weights_norm = (weights - weights.min()) / (weights.max() - weights.min())

node_sizes = [G.degree(n, weight='weight')*150 for n in G.nodes()]

plt.figure(figsize=(10, 8))
nx.draw(G, width = weights_norm, node_size=node_sizes, node_color="green", edge_color="gray", with_labels=False)
plt.title("Graph from congress.edges file ")
plt.show()
```
