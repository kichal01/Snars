```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
```

```python
#ex1


def checkC(graph):
    maxDegree = max(dict(graph.degree()).values())
    results = []

    for k in range(maxDegree + 1):
        sum_k = 0
        counter = 0

        for node in graph.nodes():
            if graph.degree(node) == k and k > 0:
                AvgNeighborDegree = 0
                
                neighbors = graph.neighbors(node)
                for n in neighbors:

                    AvgNeighborDegree += graph.degree(n)
                AvgNeighborDegree = AvgNeighborDegree/ k
                sum_k += AvgNeighborDegree
                counter += 1

        if counter != 0:
            results.append(sum_k / counter)
        else:
            results.append(0)

    return results

    
def drawResults(results, title):
    plt.figure(figsize=(10,6))
    X = np.arange(len(results))
    plt.plot(X, results, color='red')
    plt.xlabel('Node degree')
    plt.ylabel('Average degree of neighbors')
    plt.title(title)
    plt.grid()
    plt.show()



    




```

```python
G = nx.erdos_renyi_graph(100, 0.05, seed=42)

fig = plt.figure(figsize=(5,4))


pos = nx.spring_layout(G, seed=10396953)
nx.draw_networkx_nodes(G,  pos = pos, node_size=25)
nx.draw_networkx_edges(G,  pos = pos, alpha=0.6)
```

```python
res = checkC(G)
print(res)
```

```python
drawResults(res,'Erdos-Renyi Graph')
```

```python
G =nx.barabasi_albert_graph(100, 3, seed=42)


fig = plt.figure(figsize=(5,4))


pos = nx.spring_layout(G, seed=10396953)
nx.draw_networkx_nodes(G,  pos = pos, node_size=25)
nx.draw_networkx_edges(G,  pos = pos, alpha=0.6)
```

```python
res = checkC(G)
print(res)
```

```python
drawResults(res,'Barabashi-Albert Graph')
```
