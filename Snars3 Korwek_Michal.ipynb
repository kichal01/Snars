```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#exercise 1
np.random.seed(42)

X = np.random.normal(0,1,5000)
avg = np.mean(X)
std = np.std(X)


plt.hist(X, color = 'pink', bins = 30)
plt.axvline(avg, color = 'green')
plt.axvline(avg+ std, color = 'blue')
plt.axvline(avg+ 2*std, color = 'orange')
plt.axvline(avg+ 3*std, color = 'red')

plt.axvline(avg- std, color = 'blue')
plt.axvline(avg- 2*std, color = 'orange')
plt.axvline(avg- 3*std, color = 'red')
plt.title('Gaussian distribution simulation, with 1,2,3 sigma values')
plt.show()
```

```python
#exercise 2
np.random.seed(42)

X = np.random.geometric(0.1,1000)
X.sort()
top20 = X[int(-0.2*X.shape[0]):]
down80 = X[:int(0.8*X.shape[0])]

sum20 = np.sum(top20)
sum80 = np.sum(down80)

plt.hist([down80,top20], bins = np.max(X))
plt.title('Pareto check for p = 0.1')
plt.show()
print(sum20)
print(sum80)
```

```python
#exercise 3

np.random.seed(42)

X = np.random.pareto(3,5000)
avg = np.mean(X)
std = np.std(X)


plt.hist(X, color = 'pink', bins = 30)
plt.axvline(avg, color = 'green')
plt.axvline(avg+ std, color = 'blue')
plt.axvline(avg+ 2*std, color = 'orange')
plt.axvline(avg+ 3*std, color = 'red')

plt.axvline(avg- std, color = 'blue')
plt.axvline(avg- 2*std, color = 'orange')
plt.axvline(avg- 3*std, color = 'red')

plt.xlim(-2,np.max(X)+1)
plt.title('continuous power distribution simulation, with 1,2,3 sigma values')
plt.show()
```

```python
import networkx as nx
```

```python
#exercise 5

G_er = nx.erdos_renyi_graph(200, 0.04, seed=42)
degree_sequence = sorted((d for n, d in G_er.degree()), reverse=True)

avg_emp = np.mean(degree_sequence)
avg_estimator = (200-1)*0.04

std_emp = np.std(degree_sequence)
std_estimator = np.sqrt((200-1)*0.04*(1-0.04))


fig = plt.figure(figsize=(9,18))
axgrid = fig.add_gridspec(6, 5)

pos = nx.spring_layout(G_er, seed=10396953)

ax0 = fig.add_subplot(axgrid[0:3, :])
nx.draw_networkx_nodes(G_er, pos, ax=ax0,  node_size=25)
nx.draw_networkx_edges(G_er, pos, ax=ax0,  alpha=0.6)
ax0.set_title('ER graph')

ax1 = fig.add_subplot(axgrid[3:4, :])
ax1.plot(degree_sequence, "b-", marker="o")
ax1.set_title("Degree Rank Plot")
ax1.set_ylabel("Degree")
ax1.set_xlabel("Rank")

ax2 = fig.add_subplot(axgrid[5:, :])
ax2.bar(*np.unique(degree_sequence, return_counts=True))
ax2.set_title("Degree histogram")
ax2.set_xlabel("Degree")
ax2.set_ylabel("# of Nodes")

ax2.axvline(avg_emp, color = 'red')
ax2.axvline(avg_estimator, color = 'orange')


plt.show()
```

```python
print('empirical mean '+ str(avg_emp  ))
print('theoretical mean '+ str(avg_estimator  ))


print('empirical std deviation '+ str(std_emp ))
print('theoretical std deviation '+ str(std_estimator  ))
```

```python

G_er = nx.barabasi_albert_graph(200, 3, seed=42)
degree_sequence = sorted((d for n, d in G_er.degree()), reverse=True)

avg_emp = np.mean(degree_sequence)
avg_estimator = (200-3)*3*2/200

std_emp = np.std(degree_sequence)

fig = plt.figure(figsize=(9,18))
axgrid = fig.add_gridspec(6, 5)

pos = nx.spring_layout(G_er, seed=10396953)

ax0 = fig.add_subplot(axgrid[0:3, :])
nx.draw_networkx_nodes(G_er, pos, ax=ax0,  node_size=25)
nx.draw_networkx_edges(G_er, pos, ax=ax0,  alpha=0.6)
ax0.set_title('BA graph')

ax1 = fig.add_subplot(axgrid[3:4, :])
ax1.plot(degree_sequence, "b-", marker="o")
ax1.set_title("Degree Rank Plot")
ax1.set_ylabel("Degree")
ax1.set_xlabel("Rank")

ax2 = fig.add_subplot(axgrid[5:, :])
ax2.bar(*np.unique(degree_sequence, return_counts=True))
ax2.set_title("Degree histogram")
ax2.set_xlabel("Degree")
ax2.set_ylabel("# of Nodes")

ax2.axvline(avg_emp, color = 'red')
ax2.axvline(avg_estimator, color = 'orange')


plt.show()
```
